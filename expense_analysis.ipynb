```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/monthly_expenses.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
total = df["Amount"].sum()
print("Total Monthly Expense:", total)
```

```python
category_expense = df.groupby("Category")["Amount"].sum().sort_values(ascending=False)
category_expense
```

```python
category_expense.plot(kind="bar")
plt.title("Category-wise Monthly Expenses")
plt.xlabel("Category")
plt.ylabel("Amount")
plt.show()
```

```python
payment_mode = df.groupby("Payment_Mode")["Amount"].sum()
payment_mode
```

```python
daily_avg = df.groupby("Date")["Amount"].sum().mean()
print("Average Daily Expense:", daily_avg)
```

# INSIGHTS:-

1. The highest spending category is Food, indicating that daily meals are the major cost driver.
2. Total monthly expense is ₹7435, giving a clear view of overall expenditure.

4. Fixed costs like Rent and Utilities take a significant portion of the budget, which is unavoidable.
5. Non-essential spending (Shopping, Others) can be optimized to save money.
